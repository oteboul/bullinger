```python
df = agg.metrics_df

def scatter_all(df):
    fig, axes = plt.subplots(2, 3, figsize=(9*3, 7*2))
    for i, s in enumerate([1, 2]):
        for j, installation in enumerate(['all', 'avec', 'sans']):
            ax = axes[i, j]
            for ad in [True, False]:
                ddf = df[(df.ad == ad) & (df.semester == s)]
                ax.scatter(ddf['s_{}'.format(installation)],
                           ddf['r_{}'.format(installation)],
                           s=200, label='AD' if ad else 'TD', zorder=2)
            bounds = ax.get_ylim()
            ax.plot([0.0, 1.0], bounds, 'k--', alpha=0.4, zorder=1)
            ax.legend(fontsize=16)
            ax.set_title('Semestre {}, {} installation'.format(s, installation), fontsize=18)
            ax.set_xlabel('Intensit du Stimulus', fontsize=16)
            ax.set_ylabel('Facteur de Rponse', fontsize=16)
        
scatter_all(df)
```

```python
def show_baby(name, agg=agg):
    vas = agg.per_baby[name]
    k = len(vas)
    fig, axes = plt.subplots(1, k, figsize=(k*14, 10))
    for i, va in enumerate(vas):
        viewer = visualizer.AnnotationsVisualizer(va, actor_height=0.6, tags=agg.tags)
        viewer.chronogram(ax=axes[i])
    
show_baby('Carlo')
```

```python
def show_stimuli(agg):
    fig, axes = plt.subplots(2, 2, figsize=(20, 16), sharey='row')
    for i, relative in enumerate((True, False)):
        for j, s in enumerate(range(1, 3)):
            ax = axes[i, j]
            ad, v_ad = agg.average_stimulus(semester=s, relative=relative, autists=True)
            td, v_td = agg.average_stimulus(semester=s, relative=relative, autists=False)
            pd.DataFrame({'ad': ad, 'td': td}).plot.bar(rot=45, fontsize=18, ax=ax)
            ax.set_title('{} Semester {} (AD={} | TD={})'.format(
                'relative' if relative else 'absolute', s, int(v_ad), int(v_td)), fontsize=20)
    
show_stimuli(agg)
```
